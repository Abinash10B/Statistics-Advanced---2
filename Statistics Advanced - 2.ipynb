1.What is hypothesis testing in statistics?
Ans->Hypothesis testing is a statistical technique used to decide whether a given assumption about a population is true or not based on sample data. In this method, we first form a null hypothesis, which usually states that there is no effect, no difference, or no relationship between variables, and an alternative hypothesis, which states that there is an effect or a difference. After collecting sample data, a suitable statistical test is applied and a p-value is calculated, which represents the probability of obtaining the observed result purely by chance. A significance level, commonly taken as 0.05, is used as a reference for decision making. If the p-value is less than or equal to the significance level, the null hypothesis is rejected, indicating that the result is statistically significant; if the p-value is greater than the significance level, the null hypothesis is not rejected. Thus, hypothesis testing provides a systematic and scientific way to draw conclusions and make decisions using data rather than assumptions or personal judgment.

2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
Ans->Null Hypothesis and Alternative Hypothesis

In statistics, hypothesis testing is a formal procedure used to make decisions about a population based on sample data. The two main types of hypotheses used in this process are the null hypothesis and the alternative hypothesis, which are set before conducting any statistical test.

The null hypothesis, denoted as H₀, is the initial or default assumption. It states that there is no effect, no difference, or no relationship between the variables being studied. In other words, it assumes that any observed change or difference in the sample data has occurred purely due to random chance. The null hypothesis is always clearly stated and is the hypothesis that the statistical test attempts to evaluate. For example, if a company claims that a new teaching method improves student performance, the null hypothesis would state that the new teaching method has no effect on student performance.

The alternative hypothesis, denoted as H₁ or Hₐ, is the opposite of the null hypothesis. It states that there is an effect, a difference, or a relationship between variables. This hypothesis represents the claim or statement that the researcher wants to prove using statistical evidence. Continuing the previous example, the alternative hypothesis would state that the new teaching method does improve student performance. The alternative hypothesis can be one-tailed (showing direction, such as increase or decrease) or two-tailed (showing any difference).

The main difference between the null and alternative hypotheses lies in their purpose and interpretation. The null hypothesis assumes no change and acts as a baseline for comparison, while the alternative hypothesis suggests a meaningful change or effect. During hypothesis testing, a p-value is calculated and compared with a chosen significance level (usually 0.05). If the p-value is less than or equal to the significance level, the null hypothesis is rejected in favor of the alternative hypothesis; otherwise, the null hypothesis is not rejected.

In conclusion, the null hypothesis and alternative hypothesis are essential components of hypothesis testing. The null hypothesis provides a starting assumption, while the alternative hypothesis represents the research claim. Together, they help researchers make objective, data-based decisions rather than relying on assumptions or personal opinions.

3.Explain the significance level in hypothesis testing and its role in deciding the outcome of a test?
Ans->**Significance Level in Hypothesis Testing**

The **significance level**, denoted by **α (alpha)**, is a pre-decided value in hypothesis testing that represents the **maximum probability of making a wrong decision by rejecting a true null hypothesis**. In simple words, it shows how much risk a researcher is willing to take while concluding that a result is statistically significant. Commonly used significance levels are **0.05 (5%)**, **0.01 (1%)**, and **0.10 (10%)**, where 0.05 is the most widely accepted in practical applications.

The role of the significance level is to act as a **decision boundary** for the hypothesis test. After collecting sample data and performing a statistical test, a **p-value** is calculated, which indicates the probability of obtaining the observed result purely by chance assuming that the null hypothesis is true. This p-value is then compared with the chosen significance level. If the **p-value is less than or equal to α**, the result is considered statistically significant, and the **null hypothesis is rejected** in favor of the alternative hypothesis. On the other hand, if the **p-value is greater than α**, the null hypothesis is not rejected, meaning there is not enough evidence to support the alternative hypothesis.

Thus, the significance level plays a crucial role in controlling decision errors and ensuring consistency in hypothesis testing. By setting α in advance, researchers avoid biased decisions and maintain objectivity, making the testing process reliable and scientifically valid.

4.What are Type I and Type II errors? Give examples of each.
Ans-> **Type I and Type II Errors**

In hypothesis testing, **Type I and Type II errors** are mistakes that can occur while making decisions about hypotheses.

A **Type I error** occurs when the **null hypothesis (H₀) is rejected even though it is actually true**. This means we conclude that there is an effect or difference when in reality there is none. It is also called a **false positive**, and its probability is equal to the **significance level (α)**.
**Example:** Suppose a medical test concludes that a healthy person has a disease. In this case, the null hypothesis (person is healthy) is true, but it is rejected, leading to a Type I error.

A **Type II error** occurs when the **null hypothesis (H₀) is not rejected even though it is false**. This means we fail to detect an effect or difference that actually exists. It is also called a **false negative**, and its probability is denoted by **β (beta)**.
**Example:** Suppose a medical test fails to detect a disease in a person who actually has it. Here, the null hypothesis (person is healthy) is false, but it is not rejected, resulting in a Type II error.

In summary, a Type I error means detecting something that is not real, while a Type II error means failing to detect something that is real. Both errors are important in hypothesis testing, and reducing one often increases the chance of the other.


5.What is the difference between a Z-test and a T-test? Explain when to use each.
Ans->Difference Between Z-test and T-test

In statistics, both Z-test and T-test are hypothesis testing techniques used to determine whether there is a significant difference between sample statistics and population parameters. Although both tests serve a similar purpose, they differ in terms of assumptions, sample size, and data requirements. Understanding these differences helps in selecting the appropriate test for a given situation.

A Z-test is used when the population standard deviation (σ) is known and the sample size is large, typically n ≥ 30. It assumes that the sampling distribution of the mean follows a normal distribution or that the Central Limit Theorem applies due to the large sample size. Z-tests are commonly applied in quality control, large-scale surveys, and standardized testing where population parameters are known. The test statistic in a Z-test follows the standard normal distribution.

A T-test, on the other hand, is used when the population standard deviation is unknown and the sample size is small (usually n < 30). In such cases, the sample standard deviation is used as an estimate of the population standard deviation. The T-test assumes that the data is approximately normally distributed and uses the t-distribution, which accounts for additional uncertainty due to small sample sizes. T-tests are widely used in academic research and experiments involving limited data. Common types of T-tests include one-sample t-test, independent t-test, and paired t-test.

In conclusion, a Z-test is preferred when working with large samples and known population variance, while a T-test is more suitable for small samples with unknown population variance. Choosing the correct test ensures accurate and reliable statistical conclusions.

```python
#8.Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Simulate data from normal distribution
np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=100)

# Sample statistics
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

# 95% confidence interval
confidence_level = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
margin_error = z_critical * (std / np.sqrt(n))

lower_bound = mean - margin_error
upper_bound = mean + margin_error

print("Sample Mean:", mean)
print("95% Confidence Interval:", (lower_bound, upper_bound))

# Plot the data
plt.hist(data, bins=15)
plt.axvline(mean)
plt.axvline(lower_bound)
plt.axvline(upper_bound)
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
#9.Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean. 
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    data = np.array(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)
    z_scores = (data - mean) / std
    return z_scores

# Example dataset
dataset = [45, 50, 52, 48, 47, 51, 53, 49, 50, 54, 46, 52, 51, 48, 49]

# Calculate Z-scores
z_scores = calculate_z_scores(dataset)

print("Z-scores:", z_scores)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=10)
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.title("Histogram of Z-scores (Standardized Data)")
plt.show()
```
